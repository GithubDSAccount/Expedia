```python
#I'm only taking 100,000 rows to make it easier to play with. Obviously, we'll use the full dataset after we've explored a bit
import random
filename = 'W:/LargeDataSets/Expedia/train.csv' # I'll out the file location here to make it easier to other people to change
n = sum(1 for line in open(filename)) - 1 #number of rows in data set (not counting header)
s = 100000 #number of rows in sample datat set
skip = sorted(random.sample(xrange(1,n+1),n-s)) #the 0-indexed header will not be included in the skip list
df = pd.read_csv(filename, skiprows=skip)
```

```python
#Looks like date_time is a string, we'll have to fix that up so it's more useful to us
df["date_time"] = pd.to_datetime(df["date_time"])
df["year"] = df["date_time"].dt.year
df["month"] = df["date_time"].dt.month
#df["time"] = df["date_time"].dt.time
df.head()
```

```python
#OK, it looks like is_booking what what we're trying to predict. A lot of these categories don't seem that helpful...
```

```python
#There are still some columns that are hidden (in my display, at least)
#Let's explicity look at them to see if they are floats or not
df[df.columns[8:16]]
```

```python
#srch_ci and srch_co are strings, so let's get rid of them for the time being, just to make the df easier to manipulate
if 'srch_ci' in df: #I'm adding these if statements so if you've already dropped them you can still run this code block
    df=df.drop('srch_ci', 1)
if 'srch_co' in df:
    df=df.drop('srch_co', 1)
```

```python
#Let's get rid of the rows with NaN and see how much we lose
df1 = df.dropna()
df1.shape
```

```python
#We lost 36%, which is too many. It looks like orig_destination_distance is the cause of the problem, so let's get rid of it for now
```

```python
df2 = df.drop('orig_destination_distance', axis=1)
df2 = df2.dropna()
df2.shape
```

```python
#And it looks like orig_destination_distance was the only one with NaN, this makes it much easier
#I'll make this our base df again
df = df2
```

```python
#Let's remove is_booking from this dataframe so we can predict with it
train = df.drop('is_booking', 1)
train = train.drop('date_time', 1)
print(train.columns.values)
test = df['is_booking']
```

```python
#Let's see what percentage of people book, just to get an idea
print("{percentage}% of people book.".format(percentage=100 * sum(test)/len(test)))
```

```python
#Now let's break up this smaller data set into a training and a testing group
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train, test, test_size=0.2, random_state=585487)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=0)
tree.fit(X_train, y_train)
pred = tree.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, pred))
```

```python
#This might seem good, but I'm suspicious. With so few people booking you have to be careful
```

```python
true = 100 * sum(y_test)/len(y_test)
predicted = 100 * sum(pred)/len(pred)
print("{true}% of people in the test set booked, but your prediction was {predicted}.".format(true=true, predicted=predicted))
#print("{percentage}% of people book.".format(percentage=100 * sum(pred)/len(pred)))
```
