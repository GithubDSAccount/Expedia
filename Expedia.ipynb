```python
#Looks like date_time is a string, we'll have to fix that up so it's more useful to us
df["date_time"] = pd.to_datetime(df["date_time"])
df["year"] = df["date_time"].dt.year
df["month"] = df["date_time"].dt.month
#df["time"] = df["date_time"].dt.time
df.head()
```

```python
#Some of the columns are not show, so let's take a look at them all
col = df.columns.values
print(col)
```

```python
'''OK, it looks like is_booking what what we're trying to predict. A lot of these categories don't seem that helpful...'''
```

```python
#Let's remove is_booking from this dataframe so we can predict with it
train = df.drop('is_booking', 1)
train = train.drop('date_time', 1)
print(train.columns.values)
test = df['is_booking']
```

```python
#Let's see what percentage of people book, just to get an idea
print("{percentage}% of people book.".format(percentage=100 * sum(test)/len(test)))
```

```python
#Now let's break up this smaller data set into a training and a testing group
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train, test, test_size=0.2, random_state=585487)
```
